# Datasets: Cleaning & Visualizations (Code)

This imports the pandas and matplotlib.pyplot libraries. The pandas library which is used for data manipulation and analysis for structured datasets like CSV files. The matplotlib.pyplot library is used to create visualizations such as bar graphs and scatterplots. Plt is used as a shortened code.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

The first line loads the dataset named disneyplus_userbase.csv into a pandas DataFrame called df. The disneyplus_userbase.csv is the dataset that I downloaded from Kaggle including information about the new and active subscription dates from Disney+ users. 

The second part converts the Join Date column in the dataset to a datetime object ensuring that the dates are recognized in a proper date format. This makes it easier for analysis or filtering needed later on. The Last Payment Date column is also converted to a datetime object.

```python
# Load the disney+ dataset from Kaggle
df = pd.read_csv('disneyplus_userbase.csv')

# Convert the 2 date columns to datetime
df['Join Date'] = pd.to_datetime(df['Join Date'])
df['Last Payment Date'] = pd.to_datetime(df['Last Payment Date'])
```

This code is in charge of filtering the dataset to only include users who joined Disney+ in the year 2021. It creates a new DataFrame called joined_2021 which contains rows where the Join Date is between January 1, 2021 and December 31, 2021.

```python
# Filter for users who joined in the year 2021
joined_2021 = df[(df['Join Date'] >= '2021-01-01') & (df['Join Date'] <= '2021-12-31')]
```

This code extracts the year (yyyy) from the Last Payment Date column and creates a new column called Last Payment Year, in the joined_2021 DataFrame. This makes it easier when analyzing the data by a specific year. I had some trouble with formatting and received errors in my code. I asked ChatGPT for some help and it suggested using .loc which is used to update specific rows or columns in a DataFrame based on their labels.

```python
# Extract the year from the last payment date
joined_2021.loc[:, 'Last Payment Year'] = joined_2021['Last Payment Date'].dt.year
```

This code is responsible for counting the number of users by their last payment year and ensures that all years between 2021 and 2024 are included. For any missing years, they are filled with a 0. The reindex function allows for consistent formatting of visualizations.

```python
# Count the number of users for each last payment year
payment_counts = joined_2021['Last Payment Year'].value_counts().sort_index()

# Makes sure that all years between 2021 and 2024 are included
years = [2021, 2022, 2023, 2024]
payment_counts = payment_counts.reindex(years, fill_value=0)

print(payment_counts)
```

This code creates a bar graph that shows the total users from 2021 grouped by their last payment year (2021-2024). It labels each bar with the user count, assigns specific colors to each bar, and ensures all years are displayed on the x-axis. The purpose of this part is to see how many users that joined in 2021 (a peak year in the pandemic) are still paying for their Disney+ subscriptions. This shows how long each user decided to keep their subscription for.

```python
# Plot the bar graph: last payment year for users who joined in 2021
plt.bar(payment_counts.index, payment_counts.values, color=['red', 'blue', 'purple', 'green'])
plt.title('Last Payment Year for Users Who Joined in 2021')
plt.xlabel('Last Payment Year')
plt.ylabel('Number of Users')
plt.xticks(payment_counts.index)  # Ensure all years are labeled
for i, v in enumerate(payment_counts.values):
    plt.text(payment_counts.index[i], v + 0.5, str(v), ha='center')
plt.savefig('2021_barchart.png')
plt.show()
```

This code filters users who joined in the year 2022 and creates a separate copy of the data for additional analysis. 

```python
# Filter users who joined in the year 2022
joined_2022 = df[(df['Join Date'] >= '2022-01-01') & (df['Join Date'] <= '2022-12-31')]
```

This code extracts the year from the Last Payment Date for users who joined in the year 2022 and adds it to a new column called Last Payment Year in the dataset. I also used the .loc function again, which is used to update specific rows or columns in a DataFrame.

```python
# Extract the year from the last payment date using .loc for users who joined in 2022
joined_2022.loc[:, 'Last Payment Year'] = joined_2022['Last Payment Date'].dt.year
```

This code counts the total number of users for each Last Payment Year from those who joined in the year 2022. It ensures that all relevant years (2022-2024) are included. For any missing values, they are filled with a 0.

```python
# Count the number of users for each last payment year
payment_counts = joined_2022['Last Payment Year'].value_counts().sort_index()

# Make sure all relevant years (2022-2024) are included
years = [2022, 2023, 2024]
payment_counts = payment_counts.reindex(years, fill_value=0)
```

This code creates a bar graph that shows the total users from 2022 grouped by their last payment year (2022-2024). It labels each bar with the user count, assigns specific colors to each bar, and ensures all years are displayed on the x-axis. The purpose of this part is to see how many users that joined in 2022 (another peak year in the pandemic) are still paying for their Disney+ subscriptions. This shows how long each user decided to keep their subscription for.

```python
# Plot the bar chart: last payment year for users who joined in 2022
plt.bar(payment_counts.index, payment_counts.values, color=['orange', 'green', 'blue'])
plt.title('Last Payment Year for Users Who Joined in 2022')
plt.xlabel('Last Payment Year')
plt.ylabel('Number of Users')
plt.xticks(payment_counts.index)  # Ensure all years are labeled
for i, v in enumerate(payment_counts.values):
    plt.text(payment_counts.index[i], v + 0.5, str(v), ha='center')
plt.savefig('2022_barchart.png')
plt.show()
```

---

This code extracts the month and year from the Join Date and Last Payment Date columns in the Disney+ userbase dataset, adding them as Join Month and Last Payment Month to the DataFramer for easier monthly analysis. Before this process, dates included month, day, and year.

```python
# Extract the month and year from Join Date and Last Payment Date
df['Join Month'] = df['Join Date'].dt.to_period('M')  # Month-year period
df['Last Payment Month'] = df['Last Payment Date'].dt.to_period('M')

# View the updated DataFrame
print(df[['User ID', 'Join Date', 'Join Month', 'Last Payment Date', 'Last Payment Month']].head())
```

This code groups the data by Join Month to calculate the total number of new subscriptions for each month, creating a DataFrame called monthly_joiners. This step is important when creating a new csv file that only contains data for new subscriptions. 

```python
# Group by Join Month
monthly_joiners = df.groupby('Join Month').size().reset_index(name='New Subscriptions')

# View the result
print(monthly_joiners.head())
```

This code groups the data by Last Payment Month to calculate the total number of active payments for each month, creating a DataFrame named monthly_payments. This step is important when creating a new csv file that only contains data for active subscription payments.

```python
# Group by Last Payment Month
monthly_payments = df.groupby('Last Payment Month').size().reset_index(name='Active Payments')

# View the result
print(monthly_payments.head())
```

This code generates a bar chart to present the total number of new subscriptions for each month from (Join Month). It also sets up labels for the axes and ajusts the data visualization for better readability.

```python
# Plot monthly new subscriptions
plt.bar(monthly_joiners['Join Month'].astype(str), monthly_joiners['New Subscriptions'], color='blue')
plt.title('Monthly New Subscriptions')
plt.xlabel('Month')
plt.ylabel('Number of New Subscriptions')
plt.xticks(rotation=45)
plt.savefig('monthly_new_subs.png')
plt.show()
```

This code generates a bar chart to present the total number of active payments for each month from (Last Payment Month). It also sets up labels for the axes and adjusts the data visualization for better readability. 

```python
# Plot monthly active payments/subscriptions
plt.bar(monthly_payments['Last Payment Month'].astype(str), monthly_payments['Active Payments'], color='green')
plt.title('Monthly Active Payments')
plt.xlabel('Month')
plt.ylabel('Number of Active Payments')
plt.xticks(rotation=45)
plt.savefig('monthly_active_payments.png')
plt.show()
```

This code is for the purpose of renaming the columns in the dataset so they match. Right now, they are titled 'Join Month' and 'Last Payment Month', but the code below renames both of them to 'Month' so everything is consistent. This will be helpful for merging and comparisons.

```python
# Rename 'Join Month' to 'Month' in monthly_joiners
monthly_joiners.rename(columns={'Join Month': 'Month'}, inplace=True)

# Rename 'Last Payment Month' to 'Month' in monthly_payments
monthly_payments.rename(columns={'Last Payment Month': 'Month'}, inplace=True)
```

This code saves the monthly_joiners DataFrame to separate csv files for new subscriptions and active payments. The following csv files will be used for correlations in the next section.

```python
# Save New Subscriptions to a new csv file
monthly_joiners.to_csv('monthly_new_subscriptions.csv', index=False)

# Save Active Payments to a new csv file
monthly_payments.to_csv('monthly_active_payments.csv', index=False)
```

**COVID-19 Dataset Cleaning**

This code loads the dataset covid19_cases.csv that I downloaded from Kaggle and sourced from WHO into a DataFrame called covid_data. This step is similar to how I loaded the Disney+ userbase dataset in the previous section.

```python
# Load the COVID-19 case dataset from Kaggle/WHO
covid_data = pd.read_csv('covid19_cases.csv')

# View first few rows of the dataset
print(covid_data.head())
```

Since the datetime function uses a specific format that is different from the one that the COVID data came in, this code converts the Date_Reported column in the covid_data DataFrame. The format used is dd/mm/yyyy, compared to the original format of mm/dd/yyyy.

```python
# Use format parameter in pd.to_datetime to set the day to be first in order
covid_data['Date_Reported'] = pd.to_datetime(covid_data['Date_Reported'], format='%d/%m/%Y')

# Convert Date_Reported to datetime
covid_data['Date_Reported'] = pd.to_datetime(covid_data['Date_Reported'])

# Check/View data types
print(covid_data.dtypes)
```

Similar to the process used for the Disney+ userbase dataset, this code extracts the month and year from the Date_Reported column and creates a new Month column in the covid_data DataFrame. This is for the purpose of monthly aggregation in the upcoming data cells.

```python
# Create a Month column
covid_data['Month'] = covid_data['Date_Reported'].dt.to_period('M')

# View the updated DataFrame
print(covid_data[['Date_Reported', 'Cumulative_Cases', 'Month']].head())
```

This code sorts the covid_data DataFrame by the Date_Reported column in ascending order. 

```python
# Sort the data by Date_Reported in ascending order
covid_data = covid_data.sort_values(by=['Date_Reported'])

# View the data
print(covid_data[['Date_Reported', 'Cumulative_Cases']].head())
```

This code aggregates the cumulative COVID-19 cases by month using the Month column and also creates a new DataFrame called covid_monthly. The new DataFrame would store the results.

```python
# Aggregate cumulative cases by month
covid_monthly = covid_data.groupby('Month')['Cumulative_Cases'].sum().reset_index()

# View the monthly aggregated dataset
print(covid_monthly.head())
```

This code saves the monthly cumulative cases to a separate csv file called covid_monthly_cases.csv. This file can now be used to find relationships with the files created above from the Disney+ userbase.

```python
# Save the monthly cumulative cases to a new csv file
covid_monthly.to_csv('covid_monthly_cases.csv', index=False)
```

**Merging the Datasets**

This code loads the datasets monthly_new_subscriptions.csv and monthly_active_payments.csv into the new DataFrames called new_subs and active_subs. This will make the process of merging datasets much easier.

```python
# Load the new datasets created from the previous section
new_subs = pd.read_csv('monthly_new_subscriptions.csv')
active_subs = pd.read_csv('monthly_active_payments.csv')

# View the datasets
print(new_subs.head())
print(active_subs.head())
```

This code merges the new_subs and active_subs DataFrames on the Month column using an inner join. A new combined dataset called disney_data is also created which includes the variables: month, new subscriptions, and active payments.

```python
# Merge the two datasets on 'Month'
disney_data = pd.merge(new_subs, active_subs, on='Month', how='inner')

# View the combined dataset
print(disney_data.head())
```

This code loads the covid_monthly_cases.csv file into the covid_data DataFrame. Then it merges the Disney+ dataset (disney_data) that was created in the previous code cell, with the COVID-19 dataset (covid_data) on the Month column using an inner join. A new combined dataset called combined_data is created. It includes all the variables I need to conduct a correlation analysis: month, new subscriptions, active payments, and cumulative cases.

```python
# Load COVID-19 data
covid_data = pd.read_csv('covid_monthly_cases.csv')

# Merge Disney+ data with COVID-19 data
combined_data = pd.merge(disney_data, covid_data, on='Month', how='inner')

# View the fully combined dataset
print(combined_data.head())
```

This code calculates the correlation matrix for new subscriptions, active payments, and cumulative cases in the combined_data DataFrame. The matrix lays out the values needed to determine the strength and direction of the linear relationships between the variables.

```python
# Correlation matrix code
correlation_matrix = combined_data[['New Subscriptions', 'Active Payments', 'Cumulative_Cases']].corr()

# View the correlation matrix
print(correlation_matrix)
```

This code is responsible for several things. First, it creates scatterplots to analyze the relationships between COVID-19 cases and subscriptions. One of them is cumulative COVID-19 cases vs. new subscriptions and the other is cumulative COVID-19 cases vs. active subscriptions. I also added red trend lines to visualize the direction and strength of the correlation. Another important feature that this code implements are corresponding labels and titles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatterplot: COVID-19 cases vs. new subscriptions
plt.scatter(combined_data['Cumulative_Cases'], combined_data['New Subscriptions'], alpha=0.7, color='blue', label='Data Points')

# Fit and add the trend lines for the scatterplots
slope_new, intercept_new = np.polyfit(combined_data['Cumulative_Cases'], combined_data['New Subscriptions'], 1)
trend_line_new = slope_new * combined_data['Cumulative_Cases'] + intercept_new
plt.plot(combined_data['Cumulative_Cases'], trend_line_new, color='red', label='Trend Line')

# Add labels and title
plt.title('COVID-19 Cases vs. New Subscriptions (with Trend Line)')
plt.xlabel('Monthly COVID-19 Cases')
plt.ylabel('New Subscriptions')
plt.legend()
plt.grid(True)
plt.savefig('covid_vs_newsubs.png')
plt.show()

# Scatterplot: COVID-19 cases vs. active subscriptions
plt.scatter(combined_data['Cumulative_Cases'], combined_data['Active Payments'], alpha=0.7, color='green', label='Data Points')

# Fit and add the trend line
slope_active, intercept_active = np.polyfit(combined_data['Cumulative_Cases'], combined_data['Active Payments'], 1)
trend_line_active = slope_active * combined_data['Cumulative_Cases'] + intercept_active
plt.plot(combined_data['Cumulative_Cases'], trend_line_active, color='red', label='Trend Line')

# Add labels and title
plt.title('COVID-19 Cases vs. Active Subscriptions (with Trend Line)')
plt.xlabel('Monthly COVID-19 Cases')
plt.ylabel('Active Subscriptions')
plt.legend()
plt.grid(True)
plt.savefig('covid_vs_activesubs.png')
plt.show()
```

This means that there was a drop of around 22.3% subscribed users between those two years.

I was surprised to find that there were conflicting results between the two correlations. For cumulative COVID-19 cases vs. new subscriptions, the resulting correlation coefficient was r=-0.335, which is considered to be a weak negative correlation. The range for correlation coefficients is -1 to 1 with -1 representing a perfect negative correlation and 1 representing a perfect positive correlation. Since there was a weak negative correlation for cases and new subscriptions, this provides insight into the fact that months with higher cases saw a slight decrease in new subscriptions and new subscriptions were less impacted by COVID-19 trends. Factors such as high unemployment rates or financial insecurity during the pandemic might have led to price sensitivity as subscription costs would be another outlet for users' income. This relates back to a study that I explored in my background and related work section as 'subscription fatigue' has become more relevant over the years, especially after the pandemic has started to die down. In another article that I found, [Streaming Fatigue: 39% Cancel Subscriptions and 55% Join New Services](https://www.reviews.org/tv-service/streaming-habits/), the top two reasons for the most subscription cancellations in 2023 were the need to cut back on monthly expenses and not watching enough shows. I also created a bar chart in my code section where I graphed the number of new subscriptions per month between 2020 and 2024, and it showed that the peak number of highest subscriptions for Disney+ was mid-2021 while over time, they have started to gradually decrease. 

On the other hand, for cumulative COVID-19 cases vs. active subscriptions, the resulting correlation coefficient was r=0.682, which is considered to almost be a strong correlation (typically, a strong correlation coefficient starts from 0.7). This leads to the relationship that higher COVID-19 cases were associated with more active payments. A suggestion that branches off of this finding is that existing subscribers were more likely to retain their subscriptions during peak pandemic periods as lockdowns increased streaming service demands.
